```python
import pandas as pd
import numpy as np
import copy
```

```python
df=pd.read_csv('classified_thi_landuse.csv', encoding='utf-8')
```

```python
df.head(5)
```

```python
df1=df.pivot(index='ELEMENTID', columns='CLASS_NAME', values='AREA')
#convert table and group them by ELEMENTID
```

```python
df1.head(5)
```

```python
df1=df1.fillna(0)
```

```python
df1['不渗透率']=(df1['buildings']+df1['ground']+0.7*df1['mainroad']+0.7*df1['subroad']+0*df1['greens'])/(df1['buildings']+df1['ground']+df1['mainroad']+df1['subroad']+df1['greens'])
```

```python
df1.head(5)
```

```python
疑问：为什么不渗透率分子分母都没有river,只计算陆地的？
为什么绿地的不渗透率取0， 则渗透率100%
```

```python
df1['不渗透蓄水']=(2.54*df1['buildings']+2.54*df1['ground']+3*df1['mainroad']+3*df1['subroad'])/(df1['buildings']+df1['ground']+df1['greens']+df1['mainroad']+df1['subroad'])
```

```python
df1['渗透蓄水']=(7.6*df1['greens'])/(df1['buildings']+df1['greens']+df1['ground']+df1['mainroad']+df1['subroad'])
```

```python
df1.head(5)
```

```python
df1['曼宁系数(不透水)']=(0.012*df1['buildings']+0.012*df1['ground']+0.012*df1['mainroad']+0.012*df1['subroad'])/(df1['buildings']+df1['ground']+df1['mainroad']+df1['subroad'])
```

```python
df1['曼宁系数(透水)']=0.8*df1['greens']/df1['greens']
```
